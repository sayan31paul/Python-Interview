```python
import sqlite3
```

# Create database

```python
# student database01 will be created here

#First connection object
db=sqlite3.connect("student_database01.db")  #Creates the database if it doesnot exists or connects to it if it exists.
#The database will be created in homepage.
#db is the connection object
```

```python
#create cursor object from connection
cur=db.cursor() #we are running SQL codes in python. So we have to use this.
```

# Create table in student database

```python
cur.execute("create table student(id int primary key, name text, marks int)") #creating the table
```

# Insert data into table

```python
cur.execute("insert into student(id,name,marks) values(1,'John',90)") #inserting values inside table
#error is coming because same primary key cannot be entered twice
```

```python
db.commit;  #saves the data into the table permanently.
```
