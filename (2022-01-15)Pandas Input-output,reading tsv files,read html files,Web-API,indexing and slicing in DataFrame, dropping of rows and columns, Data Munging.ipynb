Dropping of rows and columns in pandas

```python
df1
```

```python
df1.columns
```

```python
#create a new table inside variable df2 after deleting some rows from source table
df2=df1.drop([14,15,16,17],axis=0) #axis=0 indicates rows
df2
```

```python
df1 #no changes in df1
```

```python
#dropping rows permanently using inplace=True
df1.drop([1,2,3],inplace=True,axis=0)       #axis=0 indicates rows, axis=1 indicates columns    
df1
```

```python
#dropping a column with no permanent chnages in source column but storing the new column in a new variable
df3=df1.drop('Name',axis=1)  #axis=1 indicates columns
df3
```

```python
#making permanent chnages in df1
df1.drop(['Age','Salary'],axis=1,inplace=True) #inplace=True makes the effect permanent
```
