```python
class Fraction:
    def __init__(self,n,d): #Here num and den are instance variables.
        self.num = n
        self.den = d
    
    def __str__(self): # This magic method is called automatically whenever we put an ojbect of that class under print() 
        return "{}/{}".format(self.num,self.den)
    
    #__add__ gets trigerred automatically whenever when "+" is encountered
    def __add__(self,other):#Since this is a binary operator, it needs two objects to operate. x is given to self and y is given to other.
        temp_num = self.num * other.den + other.num * self.den
        temp_den = self.den * other.den
        return "{}/{}".format(temp_num,temp_den)
    
    #__sub__ gets trigerred automatically whenever "-" is encouneterd.
    def __sub__(self,other):#Since this is a binary operator, it needs two objects to operate. x is given to self and y is given to other.
        temp_num = self.num * other.den - other.num * self.den
        temp_den = self.den * other.den
        return "{}/{}".format(temp_num,temp_den)
    
    def __mul__(self,other):
        temp_num = self.num * other.num
        temp_den = self.den * other.den
        return "{}/{}".format(temp_num,temp_den)
    
    def __truediv__(self,other):
        temp_num = self.num * other.den
        temp_den = self.den * other.num
        return "{}/{}".format(temp_num,temp_den)
```

#### Super Keyword

```python
class Phone():
    def __init__(self,price,brand,camera):
        print("Inside Phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera
        
    def buy(self):
        print("Buying Phone")
        
class SmartPhone(Phone):
    def buy(self):
        print("Buying  a smartphone")
        super().buy() #using super keyword, we are invoking the functions of the parent with the same name
        
s = SmartPhone(20000,"Apple" , 13)
s.buy()

#By super() methods we can access only parent methods and parent constructor
#We cannot access attributes using super mthods
#When ever super keyword is being called, it should be the first statement within the function.
```

```python
class Phone():
    def __init__(self,price,brand,camera):
        print("Inside Phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera
        
    def buy(self):
        print("Buying Phone")
        
class SmartPhone(Phone):
    def __init__(self,price,brand,camera,os,ram):
        super().__init__(price,brand,camera) #calling the parent constructor
        self.os = os
        self.ram = ram
        print("Inside Smartphone constructor")
        
s = SmartPhone(20000,"Samsung" , 12, "Android" , 2)

print(s.os)
print(s.brand)
```

# Types of inheritance

#### Multiple inheritance

```python
class Phone:
    def __init__(self,price,brand,camera):
        print("Inside Phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera
        
    def buy(self):
        print("Buying Phone")
        
    def return_phone(self):
        print("Returning a phone")
        
class Product:
    def review(self):
        print("Customer review")
        
    def buy(self):
        print("Buying Product")
        
class SmartPhone(Phone,Product):
    pass

#SmartPhone will give more priority to the methods in Phone followed by Product and so on.
#This is known as method resolution order.
```

```python
s = SmartPhone(20000,"Samsung" , 12)
```

```python
print(s.buy())
print(s.review())
```

```python
s.buy()
```
